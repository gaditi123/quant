```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import warnings
warnings.simplefilter('ignore')
import pandas_datareader as web
```

```python
data = web.get_data_yahoo("NFLX",start = "2020-01-01",end = "2021-03-01")
```

```python
ax =data['Volume'].groupby(data.index).sum().plot(figsize=(15,7), color='black')

plt.ylabel('Volume traded in number of contracts',fontsize='x-large')
plt.title(' Price vs volume',fontsize ='xx-large')

data['Close'].plot(ax=ax, secondary_y=True,color='b')
plt.ylabel('Close prices Nifty',fontsize='x-large')
plt.legend()
plt.show()
```
